```python
import pandas as pd
import numpy as np
from cmapPy.pandasGEXpress import parse
```

```python
data=pd.read_csv('../data/LINCS/GSE92742/GSE92742_Broad_LINCS_sig_info.txt',
                    sep='\t',header=0,index_col=0,low_memory=False)
```

```python
fil=data['pert_type']=='trt_sh.'
data=data[fil]
```

```python
genes=list(set(data['pert_iname']))
```

```python
gene_ids=pd.read_csv('../data/LINCS/GSE92742/GSE92742_Broad_LINCS_gene_info.txt',
                     sep='\t',header=0,index_col=None)
```

```python
gene_ids=gene_ids[gene_ids['pr_is_lm']==1]['pr_gene_id'].values.astype(str)
```

```python
results=pd.DataFrame(index=genes,columns=gene_ids)
```

```python
from scipy.stats import spearmanr as scor
def calc_MODZ(data):
    """calculates MODZ based on the original CMAP/L1000 study
    use only lm genes for MODZ calculation! Uses LM_GENES global
    variable."""
    if len(data)==1:
        return data.iloc[0]
    if len(data)==2:
        return np.mean(data,0)
    else:
        CM=scor(data.T)[0]
        fil=CM<0
        CM[fil]=0.01
        weights=np.sum(CM,1)-1
        weights=weights/np.sum(weights)
        weights=weights.reshape((-1,1))
        return pd.Series(np.dot(data.T,weights).reshape((-1,1)[0]),index=data.columns)
```

```python
for gene in genes[:10]:
    i=genes.index(gene)
    if i %10 == 0:
        print(i)
    sig_ids=data.index[data['pert_iname']==gene]
    signature=parse.parse('../data/LINCS/GSE92742/GSE92742_Broad_LINCS_Level5_COMPZ.MODZ_n473647x12328.gctx',
                            cid=sig_ids,rid=gene_ids).data_df.T[gene_ids]
    results.loc[gene]=calc_MODZ(signature)
    
```
